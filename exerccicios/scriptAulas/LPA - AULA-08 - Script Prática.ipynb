# Seção 3: Operadores lógicos

Utilizados quando precisamos realizar operações sobre um ou dois valores booleanos. O resultado de uma expressão lógica é um valor booleano (Verdadeiro ou Falso / True or False)

```python
#Criação de duas variáveis para comparação
numero_1 = 22
numero_2 = 16
```

```python
#Exemplo de operador and - "E"
if numero_1 > 20 and numero_2 < 18:
    print("As Duas condições são verdadeiras")
```

```python
#Exemplo de operador or - "OU"
if numero_1 > 20 or numero_2 <= 12:
    print("Uma ou duas das condições são verdadeiras")
```

```python
#Exemplo de operador not
if not (numero_1 < 30 and numero_2 < 18):
    print('Inverte o resultado da condição entre os parânteses')
```

# Seção 4: Operadores de atribuição

#### Os operadores de atribuição retornam o valor do objeto especificado pelo operando esquerdo após a atribuição, baseado no valor do operando à direita.
* São muito importante dentro do conceito de criação de variáveis, pois são responsáveis por atribuir valores às variáveis criadas

```python
#Operador de atribuição: igual 
z = 10
print(z)
```

```python
#Operador de soma (equivalente a z=z+10)
z+=10
print(z)
```

```python
#Operador de multiplicação (equivalente a z=z*10)
z*=10
print(z)
```

```python
#Operador de divisão (equivalente a z=z/10)
z/=10
print(z)
```

```python
#Operador de potência (equivalente a z=z**10)
z**=10
print(z)
```
