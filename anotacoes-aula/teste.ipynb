```python
import pandas as pd

# Carregar os dados do arquivo CSV em um DataFrame
df = pd.read_csv('dados.csv')

# Exibir as primeiras linhas do DataFrame
print("Primeiras linhas do DataFrame:")
print(df.head())

# Exibir informações sobre o DataFrame
print("\nInformações sobre o DataFrame:")
print(df.info())

# Calcular estatísticas básicas das colunas numéricas
print("\nEstatísticas básicas das colunas numéricas:")
print(df.describe())

# Filtrar o DataFrame por idade maior que 25
filtro_idade = df['Idade'] > 25
print("\nPessoas com idade maior que 25:")
print(df[filtro_idade])

# Agrupar os dados por cidade e calcular a média da idade
print("\nMédia de idade por cidade:")
print(df.groupby('Cidade')['Idade'].mean())


# Média
# Criar um filtro para selecionar apenas as linhas onde a cidade é "São Paulo"
filtro_cidade = df['Cidade'] == "São Paulo"

# Calcular a quantidade de pessoas que moram em São Paulo
quantidade_sao_paulo = filtro_cidade.sum()

# Mostrar a quantidade de pessoas que moram em São Paulo
print("Quantidade de pessoas que moram em São Paulo:", quantidade_sao_paulo)

print(df[filtro_cidade])


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV em um DataFrame
df = pd.read_csv('dados.csv')

# Contagem de pessoas por cidade
contagem_cidade = df['Cidade'].value_counts()

# Plotar o gráfico de barras
plt.figure(figsize=(8, 6))
contagem_cidade.plot(kind='bar', color='skyblue')
plt.title('Contagem de Pessoas por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Número de Pessoas')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor visualização
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adicionar linhas de grade horizontal
plt.tight_layout()  # Ajustar layout para evitar corte de rótulos
plt.show()
```
